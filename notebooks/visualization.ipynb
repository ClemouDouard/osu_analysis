# Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_withdiff = pd.read_csv('../data/interim/withdiff.csv', index_col='beatmapset_id')
df_withdiff.head()
```

```python
df_withdiff.describe()
```

```python
df_withdiff.columns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_withdiff, x='difficultyrating', hue='language_id', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Language')
plt.show()
```

```python
artist_count = df_withdiff['artist'].value_counts()
top_artist = artist_count.nlargest(6).index

df_top_artists = df_withdiff[df_withdiff['artist'].isin(top_artist)]

plt.figure(figsize=(10, 6))
sns.histplot(data=df_top_artists, x='difficultyrating', hue='artist', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Artist')
plt.show()
```

```python
creator_count = df_withdiff['creator'].value_counts()
top_creator = creator_count.nlargest(6).index

df_top_creator = df_withdiff[df_withdiff['creator'].isin(top_creator)]

plt.figure(figsize=(10, 6))
sns.histplot(data=df_top_creator, x='difficultyrating', hue='creator', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Creator')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_withdiff, x='difficultyrating', hue='genre_id', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Genre')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_withdiff, x='difficultyrating', hue='storyboard', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Storyboard')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_withdiff, x='difficultyrating', hue='video', multiple='stack', bins=10, palette='pastel')
plt.title('Distribution of Difficulty Ratings by Video')
plt.show()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_withdiff.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
