# Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_withdiff = pd.read_csv('../data/interim/withdiff.csv', index_col='beatmapset_id')
df_withdiff.head()
```

```python
df_withdiff.columns
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_withdiff.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
